```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import pandas as pd
import numpy as np
```

```python
url0 = 'https://www.senyak.am/vacharq'
apparts = []
```

```python
url = url0
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
results = soup.find_all('div', attrs={'class':'listing-item'})
```

```python
for row in results:
    app = {}
    try:
        app['id'] = row.find('h2', attrs={'class': 'h4'}).text
    except:
        app['id'] = row.find('h2', attrs={'class': 'h4'})
    try:
        app['sqm'] = row.find("ul", attrs={'class':'listing-details text-nowrap'}).text
    except:
        app['sqm'] = row.find("ul", attrs={'class':'listing-details text-nowrap'})
    try:
        app['address'] = row.find('a', attrs={'class':'listing-address popup-with-zoom-anim'}).text.replace('\n',"")
    except:
        app['address'] = row.find('a', attrs={'class':'listing-address popup-with-zoom-anim'})
    try:
        app['info'] = row.find('div', attrs={'class': 'listing-body'}).text
    except:
        app['info'] = row.find('div', attrs={'class': 'listing-body'})
    try:
        app['price'] = row.find('div', attrs={'class': 'listing-img-content'}).text.replace('\n',"")
    except:
        app['price'] = row.find('div', attrs={'class': 'listing-img-content'})

    apparts.append(app)
```

```python
for i in range(2, 66):
    url = f"{url0}?page={i}"

    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find_all('div', attrs={'class':'listing-item'})

    for row in results:
        app = {}
        try:
            app['id'] = row.find('h2', attrs={'class': 'h4'}).text
        except:
            app['id'] = row.find('h2', attrs={'class': 'h4'})
        try:
            app['sqm'] = row.find("ul", attrs={'class':'listing-details text-nowrap'}).text
        except:
            app['sqm'] = row.find("ul", attrs={'class':'listing-details text-nowrap'})
        try:
            app['address'] = row.find('a', attrs={'class':'listing-address popup-with-zoom-anim'}).text.replace('\n',"")
        except:
            app['address'] = row.find('a', attrs={'class':'listing-address popup-with-zoom-anim'})
        try:
            app['info'] = row.find('div', attrs={'class': 'listing-body'}).text
        except:
            app['info'] = row.find('div', attrs={'class': 'listing-body'})
        try:
            app['price'] = row.find('div', attrs={'class': 'listing-img-content'}).text.replace('\n',"")
        except:
            app['price'] = row.find('div', attrs={'class': 'listing-img-content'})

        apparts.append(app)
```

```python
df = pd.DataFrame.from_records(apparts)
```
