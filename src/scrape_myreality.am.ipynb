```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import pandas as pd
import numpy as np
```

```python
url0 = 'https://myrealty.am/hy/bnakaranneri-vacharq/7784'
apparts = []
```

```python
for i in range(2, 53):
    url = f"{url0}+'?page={i}"

    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find_all('div', attrs={'class':'item'})
    for row in results:
        app = {}
        try:
            app['sqm'] = row.find("div", attrs={'class':'item-sqm'}).text.replace('\n',"")
        except:
            app['sqm'] = row.find("div", attrs={'class':'item-sqm'})
        try:
            app['address'] = row.find('div', attrs={'class':'item-address'}).text.replace('\n',"")
        except:
            app['address'] = row.find('div', attrs={'class':'item-address'})
        try:
            app['rooms'] = row.find('div', attrs={'class': 'item-rooms'}).text.replace('\n',"")
        except:
            app['rooms'] = row.find('div', attrs={'class': 'item-rooms'})
        try:
            app['info'] = row.find('div', attrs={'class': 'item-short-text-info mt-auto show-on-list'})
        except:
            app['info'] = ' '
        try:
            app['floor'] = row.find('div', attrs={'class': 'item-floor'}).text.replace('\n',"")
        except:
            app['floor'] = row.find('div', attrs={'class': 'item-floor'})
        try:
            app['price'] = row.find('div', attrs={'class': 'col item-price'}).text.replace('\n',"")
        except:
            app['price'] = row.find('div', attrs={'class': 'col item-price'})
        try:
            app['type'] = row.find('div', attrs={'class': 'col-auto item-property-type d-flex align-items-center'}).text.replace('\n', '')
        except:
            app['type'] = row.find('div', attrs={'class': 'col-auto item-property-type d-flex align-items-center'})
        try:
            app['date'] = row.find('div', attrs={'class': 'item-date show-on-list d-none d-lg-block'})
        except:
            app['date'] = ' '
        apparts.append(app)
df = pd.DataFrame.from_records(apparts)
```
