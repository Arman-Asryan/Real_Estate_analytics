```python
for i in range(2, 21):
    url = f"{url0}?page={i}"

    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find_all('div', attrs={'class':'card mb-3 p-2 rounded-0 estate-container'})
    
    for row in results:
        app = {}
        try:
            app['sqm'] = row.find("div", attrs={'class':'card__item-sqm'}).get_text().replace('\n',"")
        except:
            app['sqm'] = row.find("div", attrs={'class':'card__item-sqm'})
        try:
            app['address'] = row.find('div', attrs={'class':'card-body__address text-uppercase text-truncate border-bottom py-2'}).text.replace('\n',"")
        except:
            app['address'] = row.find('div', attrs={'class':'card-body__address text-uppercase text-truncate border-bottom py-2'})
        try:
            app['rooms'] = row.find('div', attrs={'class': 'card__item-room'}).get_text().replace('\n',"")
        except:
            app['rooms'] = row.find('div', attrs={'class': 'card__item-room'})
        try:
            app['floor'] = row.find('div', attrs={'class': 'card__item-floor'}).text.replace('\n',"")
        except:
            app['floor'] = row.find('div', attrs={'class': 'card__item-floor'})
        try:
            app['price'] = row.find('div', attrs={'class': 'lead text-primary font-weight-bold border-bottom'}).text.replace('\n',"")
        except:
            app['price'] = row.find('div', attrs={'class': 'lead text-primary font-weight-bold border-bottom'})
        try:
            app['type'] = row.find('div', attrs={'class': 'card-body__building-type text-primary font-weight-600'}).text.replace('\n', '')
        except:
            app['type'] = row.find('div', attrs={'class': 'card-body__building-type text-primary font-weight-600'})

        apparts.append(app)
```

```python
df = pd.DataFrame.from_records(apparts)
```

```python
df.rooms = df.rooms.astype(str)
df.floor = df.floor.astype(str)
df.price = df.price.astype(str)
```
